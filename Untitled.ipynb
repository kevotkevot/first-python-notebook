```python
merged = pandas.merge(pot, contribs, on="calaccess_committee_id")
```

```python
merged.sort_values("amount", ascending=False)
```

```python
merged.amount.sum()
```

```python
merged.sort_values(['contributor_lastname', 'amount'])
```

```python
support = merged[merged.committee_position == 'SUPPORT']
```

```python
oppose = merged[merged.committee_position == 'OPPOSE']
```

```python
len(support)
```

```python
merged.groupby(["contributor_lastname", "contributor_firstname", "committee_position"]).amount.sum().reset_index().sort_values("amount",ascending=False)
```

```python
merged.groupby(["contributor_employer"]).amount.sum().reset_index().sort_values("contributor_employer")
```

```python
top_supporters = support.groupby(['contributor_lastname' ,'contributor_firstname']).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
chart = top_supporters.amount.plot.barh()
chart.set_yticklabels(top_supporters.contributor_lastname)
```

```python
top_supporters["contributor_fullname"] = top_supporters.contributor_firstname+ " " + top_supporters.contributor_lastname 
```

```python
top_supporters
```

```python
chart = top_supporters.amount.plot.barh()
chart.set_yticklabels(top_supporters.contributor_fullname)
```

```python
top_supporters.to_csv("top-supporters.csv",index=False)
```
